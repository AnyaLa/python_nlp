```python
df = pd.read_csv('articles.csv', index_col=0)
df
```

```python
df_sorted = df.sort_values(['date'])
df_sorted
```

```python
print('Все навостные статьи из данного корпуса опубликованы с %s по %s'%(df['date'].min(), df['date'].max()))
```

Хотим посмотреть только статьи, которые были опубликованны за последний год

```python
df[df['year'] == df['year'].max()]
```

А теперь хотим, только за последний месяц последнего года

```python
#первый способ
df_final_year = df[df['year'] == df['year'].max()]
df_final_year_month = df_final_year[df_final_year['month'] == df_final_year['month'].max()]
df_final_year_month
```

```python
#второй способ
df[(df['year'] == df['year'].max())&(df['month'] == df['month'].max())]
```
